```python
#Psychological Questionnaire Toolkit
#Developed by Arijit Bhattacharya (2026).
#Psychological Assessment Toolkit: A modular Python system for administering psychological questionnaires.
#https://github.com/BhattacharyaArijit/psychology-assessment-toolkit

import tkinter as tk
from tkinter import ttk, messagebox
import pandas as pd
import os
from datetime import datetime

OUTPUT_DIR = "./output/Participant_Responses"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# QUESTION BANKS

PANAS = [
    "1. Interested", "2. Distressed", "3. Excited", "4. Upset", "5. Strong",
    "6. Guilty", "7. Scared", "8. Hostile", "9. Enthusiastic", "10. Proud",
    "11. Irritable", "12. Alert", "13. Ashamed", "14. Inspired", "15. Nervous",
    "16. Determined", "17. Attentive", "18. Jittery", "19. Active", "20. Afraid"
]

BDI = [
    ("I do not feel sad", "I feel sad", "I am sad all the time", "I am so sad I can't stand it"),
    ("I am not discouraged", "I feel discouraged", "I have nothing to look forward", "Future is hopeless"),
    ("I do not feel like a failure", "I failed more than others", "Life full of failures", "I am a complete failure"),
    ("I enjoy as before", "I don't enjoy", "I don't get satisfaction", "Bored with everything"),
    ("I don't feel guilty", "Guilty part of time", "Guilty most time", "Guilty all the time"),
    ("I don't feel punished", "I may be punished", "Expect to be punished", "Being punished now"),
    ("Not disappointed", "Disappointed", "Disgusted", "Hate myself"),
    ("Not worse than others", "Critical of weaknesses", "Blame myself always", "Blame myself for everything"),
    ("No suicidal thoughts", "Thoughts but won't do", "Want to", "Would if chance"),
    ("Cry as usual", "Cry more", "Cry all the time", "Can't cry anymore"),
    ("Not more irritated", "Slightly more", "Quite annoyed", "Irritated all the time"),
    ("Not lost interest", "Less interest", "Lost most", "Lost all interest"),
    ("Make decisions well", "Delay decisions", "Hard to decide", "Can't decide"),
    ("Don't look worse", "Worried appearance", "Unattractive", "Ugly"),
    ("Work as well", "Extra effort", "Push myself", "Can't work"),
    ("Sleep well", "Not well", "Wake early", "Wake hours early"),
    ("Not tired", "More tired", "Tired from anything", "Too tired"),
    ("Appetite normal", "Not as good", "Much worse", "No appetite"),
    ("No weight loss", ">5 lbs", ">10 lbs", ">15 lbs"),
    ("No health worry", "Worried aches", "Very worried", "Only thing I think"),
    ("No change in sex interest", "Less interest", "Almost none", "Lost completely")
]

MUSIC_PERCEP = [
"Can you distinguish different rhythms in music?",
"Can you follow the melody in music?",
"Can you hear differences in musical tone?",
"Can you recognize words in songs?",
"Can you recognize sounds of instruments?",
"Can you hear meaning/emotion of music?",
"Can you hear music without effort?",
"Can you recognize familiar music?",
"Can you judge performance quality?",
"Do you feel confident hearing like others?",
"Does music sound in tune?"
]

MUSIC_ENGAGE = [
"Do you enjoy music in noisy environments?",
"Do you enjoy music on TV/DVD/computer?",
"Do you choose background music?",
"Do you listen while traveling?",
"Do you choose new music?",
"Do you attend public music events?",
"Do you sing, play, whistle?"
]

MUSIC_PERCEP_IMPORTANCE = [
"How important is it for you to distinguish different rhythms in music?",
"How important is it for you to follow the melody in music?",
"How important is it for you to hear differences in musical tone?",
"How important is it for you to recognise the words in songs?",
"How important is it for you to recognise sounds of instruments?",
"How important is it for you to hear the meaning/emotion of music?",
"How important is it for you to hear music without effort?",
"How important is it for you to recognise familiar music?",
"How important is it for you to judge performance quality?",
"How important is it for you to feel confident that you hear like others?",
"How important is it for you to hear music in tune?"
]

MUSIC_ENGAGE_IMPORTANCE = [
"How important is it to enjoy music in noisy environments?",
"How important is it to enjoy music on media?",
"How important is background music?",
"How important is music while travelling?",
"How important is it to explore new music?",
"How important are public music events?",
"How important is singing/playing instruments?"
]

STAI_STATE_ITEMS = [
"I feel calm","I feel secure","I am tense","I feel strained","I feel at ease","I feel upset",
"I worry about misfortunes","I feel satisfied","I feel frightened","I feel comfortable",
"I feel self-confident","I feel nervous","I am jittery","I feel indecisive","I am relaxed",
"I feel content","I am worried","I feel confused","I feel steady","I feel pleasant"
]

STAI_STATE_REVERSE = {
"I feel calm","I feel secure","I feel at ease","I feel satisfied","I feel comfortable",
"I feel self-confident","I am relaxed","I feel content","I feel steady","I feel pleasant"
}

STAI_TRAIT_ITEMS = [
"I feel pleasant","I feel nervous and restless","I feel satisfied with myself",
"I wish I could be happy as others","I feel like a failure","I feel rested","I am calm",
"Difficulties piling up","I worry too much","I am happy","I have disturbing thoughts",
"I lack self-confidence","I feel secure","I make decisions easily","I feel inadequate",
"I am content","Thoughts bother me","I can't put disappointments away","I am steady",
"Tension thinking current concerns"
]

STAI_TRAIT_REVERSE = {
"I feel pleasant","I feel satisfied with myself","I feel rested","I am calm","I am happy",
"I feel secure","I make decisions easily","I am content","I am steady"
}

NEURO_ITEMS = [
"I get stressed easily","I am relaxed most of the time","I worry about things",
"I seldom feel blue","I am easily disturbed","I get upset easily","I change my mood a lot",
"I have frequent mood swings","I get irritated easily","I often feel blue"
]

NEURO_REVERSE = {
"I am relaxed most of the time",
"I seldom feel blue"
}


# REVERSE SCORING

def reverse_score(value, min_val, max_val):
    return max_val + min_val - value



# GUI CLASS

class FormGUI:
    def __init__(self, root):
        self.root = root
        root.title("Psychological Questionnaire")
        root.geometry("1200x800")

        ttk.Label(root, text="Psychological Questionnaire", font=("Arial", 18, "bold")).pack(pady=10)

        top = ttk.Frame(root)
        top.pack()

        ttk.Label(top, text="Subject ID:", font=("Arial", 12)).grid(row=0, column=0)
        self.sid = ttk.Entry(top, width=20)
        self.sid.grid(row=0, column=1, pady=5, padx=5)

        ttk.Label(top, text="Date:", font=("Arial", 12)).grid(row=0, column=2)
        self.date = ttk.Entry(top, width=12)
        self.date.insert(0, datetime.now().strftime("%Y-%m-%d"))
        self.date.grid(row=0, column=3, pady=5)

        self.responses = {}

        self.nb = ttk.Notebook(root)
        self.nb.pack(fill="both", expand=True)
        
#Questionnaire Pages

#PANAS-SF
        self.make_page(
            "PANAS (1–5 scale)", PANAS, [1,2,3,4,5],
            "Positive and Negative Affect Schedule (PANAS-SF)\n"
            "Instruction: Indicate the extent you have felt each emotion over the past week.\n"
            "Use the scale below (in boxes):\n"
            "1 = Very Slightly or Not at all, 2 = A little, 3 = Moderately, 4 = Quite a bit, 5 = Extremely"
        )

#BDI

        self.make_BDI_page()


#MUSIC QoL PART I
        self.make_page(
            "Music QoL – Part I", MUSIC_PERCEP, [1,2,3,4,5,"N/A"],
            "Music-Related Quality of Life Questionnaire — Part I\n"
            "Instruction: This part of the questionnaire asks you about your music listening abilities,\n"
            "attitudes towards music and musical activities.\n"
            "Use the scale: 1 = Never, 2 = Rarely, 3 = Occasionally, 4 = Frequently, 5 = Always, N/A = not applicable."
        )

#MUSIC QoL PART II
        self.make_page(
            "Music QoL – Part II", MUSIC_PERCEP_IMPORTANCE, [1,2,3,4,5,"N/A"],
            "Music-Related Quality of Life Questionnaire — Part II\n"
            "Instruction: This part asks how important these musical abilities or attitudes are to you.\n"
            "Use the scale: 1 = Not important at all, 2 = Not very important, 3 = Somewhat important,\n"
            "4 = Very important, 5 = Extremely important, N/A = not applicable."
        )

#STAI STATE
        self.make_page(
            "STAI State (1–4)", STAI_STATE_ITEMS, [1,2,3,4],
            "State–Trait Anxiety Inventory (STAI) — Part I (State)\n"
            "Instruction: Read each statement and select the response describing how you feel right now.\n"
            "Use the scale:\n"
            "1 = Not at all, 2 = A little, 3 = Somewhat, 4 = Very much so"
        )

#STAI TRAIT
        self.make_page(
            "STAI Trait (1–4)", STAI_TRAIT_ITEMS, [1,2,3,4],
            "State–Trait Anxiety Inventory (STAI) — Part II (Trait)\n"
            "Instruction: Read each statement and choose the response describing how you generally feel.\n"
            "Use the scale:\n"
            "1 = Almost never, 2 = Sometimes, 3 = Often, 4 = Almost always"
        )
#NEUROTICISM
        self.make_page(
            "Neuroticism (0–4)", NEURO_ITEMS, [0,1,2,3,4],
            "Neuroticism Questionnaire\n"
            "Instruction: For each statement, select the response that best describes you.\n"
            "Rate using:\n"
            "0 = Not at all, 1 = A little, 2 = Somewhat, 3 = Quite a bit, 4 = Very much so"
        )
# Credits / References Page
        about = ttk.Frame(self.nb)
        self.nb.add(about, text="About / Credits")

        ttk.Label(about, text="Psychological Questionnaire Toolkit",
                  font=("Arial", 16, "bold")).pack(pady=10)

        ttk.Label(
            about,
            text=(
                "Developed by: Arijit Bhattacharya\n"
                "Ph.D. Candidate, Dept. of Psychology and Cognitive Sciences, Ashoka University\n"
                "Contact: arijit.bhattacharya_phd23@ashoka.edu.in\n"
                "This psychological questionnaire toolkit was developed using Python 3.11 with the Tkinter and Pandas libraries for GUI\n"
                "and data handling, respectively. Parts of the code structure and logic were assisted by ChatGPT-5 (OpenAI).",
            ),
            font=("Arial", 12),
        ).pack(pady=15)

        ttk.Label(about, text="References", font=("Arial", 14, "bold")).pack(pady=5)

        references_text = """
Beck, A. T., Ward, C. H., Mendelson, M., Mock, J., & Erbaugh, J. (1961).
An inventory for measuring depression. Archives of General Psychiatry, 4(6), 561–571.
https://doi.org/10.1001/archpsyc.1961.01710120031004

Dritsakis, G., Van Besouw, R. M., Kitterick, P., & Verschuur, C. A. (2017).
A music-related quality of life measure to guide music rehabilitation for adult cochlear implant users.
American Journal of Audiology, 26(3), 268–282.
https://doi.org/10.1044/2017_AJA-17-0008

Watson, D., Clark, L. A., & Tellegen, A. (1988).
Development and validation of brief measures of positive and negative affect: The PANAS scales.
Journal of Personality and Social Psychology, 54(6), 1063–1070.
https://doi.org/10.1037/0022-3514.54.6.1063

Spielberger, C. D., Gorsuch, R. L., Lushene, R., Vagg, P. R., & Jacobs, G. A. (1983).
Manual for the State-Trait Anxiety Inventory. Consulting Psychologists Press.

Goldberg, L. R. (1992).
The development of markers for the Big-Five factor structure.
Psychological Assessment, 4(1), 26–42.
https://doi.org/10.1037/1040-3590.4.1.26
"""

        credits_box = tk.Text(about, wrap="word", height=22, width=90, font=("Arial", 10))
        credits_box.insert("1.0", references_text)
        credits_box.config(state="disabled")
        credits_box.pack(padx=10, pady=10)

        ttk.Button(root, text="Submit", command=self.save).pack(pady=15)

# Page Builders
    def make_page(self, title, questions, options, instruction=None):
        frame = ttk.Frame(self.nb)
        self.nb.add(frame, text=title)

#Add instruction block at top of every page
        if instruction:
            ttk.Label(
                frame,
                text=instruction,
                font=("Arial", 12),
                justify="center",
                wraplength=1100
            ).pack(pady=10)

# Scrollable Questionnaire Body
        canvas = tk.Canvas(frame)
        scrollbar = ttk.Scrollbar(frame, orient="vertical", command=canvas.yview)
        form = ttk.Frame(canvas)

        form.bind("<Configure>", lambda e: canvas.configure(scrollregion=canvas.bbox("all")))
        canvas.create_window((0, 0), window=form, anchor="nw")
        canvas.configure(yscrollcommand=scrollbar.set)
    
        canvas.pack(side="left", fill="both", expand=True)
        scrollbar.pack(side="right", fill="y")

        row_index = 0
        for q in questions:
            ttk.Label(form, text=q, font=("Arial", 11, "bold")).grid(
                row=row_index, column=0, sticky="w", pady=4
            )

            var = tk.StringVar(value="")
            self.responses[q] = var

            for j, opt in enumerate(options):
                tk.Radiobutton(
                    form,
                    text=str(opt),
                    variable=var,
                    value=str(opt)
                ).grid(row=row_index, column=j+1, padx=16)

            row_index += 1


    def make_BDI_page(self):
        frame = ttk.Frame(self.nb)
        self.nb.add(frame, text="BDI (0–3)")
        ttk.Label(
            frame,
            text=(
                "Beck's Depression Inventory (BDI)\n"
                "Instruction: For each numbered item below, choose the statement that best describes your feelings.\n"
                "Select one option: 0, 1, 2, 3. Do not overthink your answer."
            ),
            font=("Arial", 12),
            justify="center",
            wraplength=1100
        ).pack(pady=15)
        
        canvas = tk.Canvas(frame)
        scrollbar = ttk.Scrollbar(frame, orient="vertical", command=canvas.yview)
        form = ttk.Frame(canvas)

        form.bind("<Configure>", lambda e: canvas.configure(scrollregion=canvas.bbox("all")))
        canvas.create_window((0, 0), window=form, anchor="nw")
        canvas.configure(yscrollcommand=scrollbar.set)

        canvas.pack(side="left", fill="both", expand=True)
        scrollbar.pack(side="right", fill="y")

        row_index = 0
        for idx, group in enumerate(BDI, start=1):
            ttk.Label(form, text=f"Q{idx}:", font=("Arial", 11, "bold")).grid(
                row=row_index, column=0, sticky="w", pady=4
            )
            var = tk.StringVar(value="")
            self.responses[f"BDI_Q{idx}"] = var
            for score, text in enumerate(group):
                tk.Radiobutton(form, text=f"({score}) {text}", variable=var, value=str(score)).grid(
                    row=row_index + score + 1, column=0, sticky="w", padx=30
                )
            row_index += 5

# Corrected Scoring Engine
    def calculate_scores(self):

# PANAS
        pos_words = ["Interested","Excited","Strong","Enthusiastic","Proud",
                     "Alert","Inspired","Determined","Attentive","Active"]
        neg_words = ["Distressed","Upset","Guilty","Scared","Hostile",
                     "Irritable","Ashamed","Nervous","Jittery","Afraid"]

        panas_pos = 0
        panas_neg = 0
        for q in PANAS:
            val = self.responses[q].get()
            if val.isdigit():
                if any(w in q for w in pos_words):
                    panas_pos += int(val)
                if any(w in q for w in neg_words):
                    panas_neg += int(val)

# BDI
        bdi_total = sum(int(v.get()) for k, v in self.responses.items()
                        if k.startswith("BDI_") and v.get().isdigit())

# Music QoL
        music_percep = sum(
            int(self.responses[q].get()) for q in MUSIC_PERCEP
            if self.responses[q].get().isdigit()
        )
        music_engage = sum(
            int(self.responses[q].get()) for q in MUSIC_ENGAGE
            if self.responses[q].get().isdigit()
        )

# STAI State
        stai_state_sum = 0
        for q in STAI_STATE_ITEMS:
            raw = self.responses[q].get()
            if raw.isdigit():
                val = int(raw)
                if q in STAI_STATE_REVERSE:
                    val = reverse_score(val, 1, 4)
                stai_state_sum += val

# STAI Trait
        stai_trait_sum = 0
        for q in STAI_TRAIT_ITEMS:
            raw = self.responses[q].get()
            if raw.isdigit():
                val = int(raw)
                if q in STAI_TRAIT_REVERSE:
                    val = reverse_score(val, 1, 4)
                stai_trait_sum += val

# Neuroticism
        neuro_sum = 0
        for q in NEURO_ITEMS:
            raw = self.responses[q].get()
            if raw.isdigit():
                val = int(raw)
                if q in NEURO_REVERSE:
                    val = reverse_score(val, 0, 4)
                neuro_sum += val

        return {
            "PANAS_Positive": panas_pos,
            "PANAS_Negative": panas_neg,
            "BDI": bdi_total,
            "Music_Perception_Total": music_percep,
            "Music_Engagement_Total": music_engage,
            "STAI_State": stai_state_sum,
            "STAI_Trait": stai_trait_sum,
            "Neuroticism": neuro_sum
        }

# SAVE
    def save(self):
        if self.sid.get().strip() == "":
            messagebox.showwarning("Missing", "Please enter Subject ID")
            return

        df = pd.DataFrame({
            "Question": list(self.responses.keys()),
            "Response": [v.get() for v in self.responses.values()],
            "Subject ID": self.sid.get(),
            "Date": self.date.get(),
        })

        filename = f"{self.sid.get()}_{datetime.now().strftime('%H%M%S')}.csv"
        df.to_csv(os.path.join(OUTPUT_DIR, filename), index=False)

        scores = self.calculate_scores()
        master_row = {"Subject ID": self.sid.get(), **scores}

        master_path = os.path.join(OUTPUT_DIR, "Master_Scores.csv")
        if os.path.exists(master_path):
            pd.DataFrame([master_row]).to_csv(master_path, mode="a", header=False, index=False)
        else:
            pd.DataFrame([master_row]).to_csv(master_path, index=False)

        messagebox.showinfo("Saved", f"Responses saved:\n{filename}")


# RUN GUI
root = tk.Tk()
gui = FormGUI(root)
root.mainloop()
```
